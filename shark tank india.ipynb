# percantage of compnies got investment or not.

```python
got=df[df["deal_amount"]>0.0].shape[0]
not_in=df[df["deal_amount"]==0.0].shape[0]
values=[got,not_in]
plt.pie(values,labels=["Got investment"," Not Recieved Investment"],autopct='%1.1f%%',startangle=35)
plt.title("Number of companies which got INvestment or Not -")
plt.show()
```

## shark deals -

```python
sharks = df[df["total_sharks_invested"]==5]
sharks = sharks.sort_values(by='deal_amount',ascending=False)
plt.figure(figsize=(10,7))
sns.set(style='darkgrid')
sns.barplot(data=sharks,x="brand_name",y="deal_amount",palette="viridis",color='pitcher_ask_amount')
plt.xlabel("Brand Name")
plt.ylabel("Deal Amount")
plt.title("5 Sharks Deals And Total Investment")
plt.xticks(rotation=60,ha='right')
plt.show()
```

## Highest Pitch Ask Amount:

```python
high = df[df["pitcher_ask_amount"] > 100]
plt.figure(figsize=(12, 8))
sns.barplot(x='brand_name', y='pitcher_ask_amount', data=high, hue='pitcher_ask_amount', palette='viridis')
plt.title("Highest Pitch Ask Amount:")
plt.show()
```

## Least pitch amount:

```python
least=df[df["pitcher_ask_amount"]<20]
plt.figure(figsize=(10,8))
sns.set_theme(style="dark")
sns.barplot(x="brand_name",y="pitcher_ask_amount",data=least,hue="pitcher_ask_amount",palette="viridis")
plt.title("least pitcher ask amount")
plt.show()
```

## Ask Equity and Deal Equity:

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='brand_name', y='ask_equity', data=high, hue='ask_equity', palette='dark')
plt.title("Ask Equity & Deal Equity of Highest Pitch Ask Amount Brand:")
plt.show()
```

## Each shark invested in deals:

```python
deals=[df["ashneer_deal"].sum(),df["anupam_deal"].sum(),df["aman_deal"].sum(),
       df["namita_deal"].sum(),df["vineeta_deal"].sum(),df["ghazal_deal"].sum(),
      df["peyush_deal"].sum()]
columns=["ashneer","anupam","aman","namita","vineeta","ghazal","peyush"]
plt.figure(figsize=(12,8))
sns.barplot(sharks,x=columns,y=deals,hue=deals,width=0.96)
plt.xlabel("shark Names")
plt.ylabel("deals Values")
plt.title("Sharks Invested In deals")
plt.show()
```
